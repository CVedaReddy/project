```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df= pd.read_csv("//content/sample_data/discharge.csv")
df.info()
df.head()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['type'].unique()
```

```python
df.drop(columns="type")
```

```python
df_b5=df[df['Battery']=='B0005'].copy()
df_b5.info()
print(df_b5.head())
print(df['id_cycle'].nunique())
```

```python
df_b5.sort_values(by=['id_cycle','Time'],inplace=True)
print(df_b5.sort_values)
```

```python
df_b5['dt'] = df_b5.groupby('id_cycle')['Time'].diff().fillna(0)
print(df_b5['dt'])
```

```python
# find incremental discharge capacity = current * time / 3600
df_b5['dAh'] = (df_b5['Current_measured'].abs() * df_b5['dt']) / 3600
print(df_b5['dAh'])
```

```python
# find incremental discharge energy = VIt / 3600
df_b5['dWh'] = (df_b5['Voltage_measured'].abs() * df_b5['Current_measured'].abs() * df_b5['dt']) / 3600
print(df_b5['dWh'])
```

```python
# cycle summary- B0005
summary_b5= df_b5.groupby('id_cycle').agg(Total_Ah=('dAh', 'sum'),
        Total_Wh=('dWh', 'sum'),Avg_Temp=('Temperature_measured', 'mean'),Start_Voltage=('Voltage_measured', 'first'),
        End_Voltage=('Voltage_measured', 'last'),Avg_Voltage=('Voltage_measured', 'mean'),
        Avg_Current=('Current_measured', 'mean'),cycle_duration=('Time', lambda x: x.max() - x.min()),temp_variance=('Temperature_measured','var'),current_variance=('Current_measured','var'),voltage_variance = ('Voltage_measured', 'var'))
print(summary_b5)
```

```python
summary_b5['R'] = (summary_b5['Avg_Voltage'] / summary_b5['Avg_Current']).abs()
print(summary_b5['R'])
```

```python
summary_b5 = summary_b5.reset_index()
print(summary_b5)
```

```python
capacity_check = df_b5.groupby('id_cycle')['Capacity'].nunique()
print(capacity_check)
provided_capacity = df_b5.groupby('id_cycle')['Capacity'].first().reset_index()
provided_capacity.rename(columns={'Capacity': 'Provided_Capacity_Ah'}, inplace=True)
print(provided_capacity)
```

```python
summary_b5 = summary_b5.merge(provided_capacity, on='id_cycle')

print("\n--- Cycle Summary Data ---")
print(summary_b5.head())
```

```python
#Average Temperature per Cycle plot

plt.figure(figsize=(10, 6))
plt.plot(summary_b5['id_cycle'], summary_b5['Avg_Temp'],
             marker='o', linestyle='-', color='red')
plt.title('Average Cycle Temperature for Battery B0005')
plt.xlabel('Cycle Number')
plt.ylabel('Average Temperature (Â°C)')
plt.grid(True)
min_temp = summary_b5['Avg_Temp'].min()
plt.ylim(bottom=min_temp * 0.95)
```

```python
# capacity fade
plt.figure(figsize=(10, 6))
plt.plot(summary_b5['id_cycle'], summary_b5['Provided_Capacity_Ah'], marker='o', linestyle='-')
plt.title('Capacity Fade for Battery B0005')
plt.xlabel('Cycle Number')
plt.ylabel('Discharge Capacity (Ah)')
plt.grid(True)
min_capacity = summary_b5['Provided_Capacity_Ah'].min()
plt.ylim(bottom=min_capacity * 0.95)


```

```python
# Equivalent Resistance per Cycle plot

plt.figure(figsize=(10, 6))
plt.plot(summary_b5['id_cycle'], summary_b5['R'],
             marker='o', linestyle='-', color='green')
plt.title('Equivalent Cycle Resistance for Battery B0005')
plt.xlabel('Cycle Number')
plt.ylabel('Equivalent Resistance (Ohm)')
plt.grid(True)
min_res = summary_b5['R'].min()
plt.ylim(bottom=min_res * 0.95)
```
